```python
import pandas as pd
import random
```

```python
names = [
    "Alice", "Bob", "Charlie", "David", "Eva",
    "Frank", "Grace", "Hannah", "Ian", "Julia"
]
```

```python
data = []
for name in names:
    skills = {
        "Java (%)": random.randint(50, 100),
        "C (%)": random.randint(50, 100),
        "Python (%)": random.randint(50, 100)
    }
    marks = {
        "Chemistry (marks)": random.randint(60, 100),
        "Maths (marks)": random.randint(60, 100),
        "Physics (marks)": random.randint(60, 100)
    }
    row = {"Name": name}
    row.update(skills)
    row.update(marks)
    data.append(row)
```

```python
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
#LINE PLOT
# Use student names as x-axis labels
xpoints = np.array(df['Name'])  # names
ypoints = np.array(df['Maths (marks)'])  # their marks in Maths

plt.figure(figsize=(10, 6))
plt.plot(xpoints, ypoints, marker='o', linestyle='-', color='purple')  # Customize as needed
plt.title("Maths Marks of Students")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.grid(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
x = df['Name']

plt.plot(x, df['Chemistry (marks)'], marker='o', label='Chemistry')
plt.plot(x, df['Maths (marks)'], marker='s', label='Maths')
plt.plot(x, df['Physics (marks)'], marker='^', label='Physics')

plt.title("Subject Marks Comparison")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

x = np.arange(len(df['Name']))  # number of students
width = 0.25  # width of each bar

plt.figure(figsize=(12, 6))
plt.bar(x - width, df['Chemistry (marks)'], width, label='Chemistry', color='orange')
plt.bar(x, df['Maths (marks)'], width, label='Maths', color='blue')
plt.bar(x + width, df['Physics (marks)'], width, label='Physics', color='green')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Comparison: Chemistry, Maths, Physics')
plt.xticks(x, df['Name'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
